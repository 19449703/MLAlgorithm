```python
import matplotlib.pyplot as plt
import numpy as np
import KdTree
import random

%reload_ext autoreload
%autoreload 2
```

```python
data = [(2,5), (6,1.4), (3.8,8), (8.5,3), (8,5.5), (1,1), (1.5,8.5)]
target = (2.1, 5.5)

kd = KdTree.KdTree(data, 1)
kd.PlotKdTree(target, 0, 10, 0, 10)
```

```python
k_list = kd.search_k_nearest(target, 3)
kd.PlotKdTreeWithKNearest(target, k_list, 0, 10, 0, 10)
```

```python
xmax = 25
ymax = 15
data = [(random.uniform(1, xmax), random.uniform(1, ymax)) for i in range(40)]
target = (random.uniform(1, xmax), random.uniform(1, ymax))

kd = KdTree.KdTree(data, 1)
kd.PlotKdTree(target, 0, xmax, 0, ymax, plotCoord=False)
```

```python
k_list = kd.search_k_nearest(target, 5)
kd.PlotKdTreeWithKNearest(target, k_list, 0, xmax, 0, ymax, plotCoord=False)
```
